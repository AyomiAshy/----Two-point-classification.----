```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/insurance_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Split the data into testing and training
X_train,X_test,y_train,y_test=train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
# Build the model
model=LogisticRegression()
# Train the model with the data on the x axis
model.fit(X_train,y_train)
```

```python
# Make predictions
predicted_values=model.predict(X_test)
# Overview of predicted values
predicted_values
```

```python
# Compare the predicted values and the actual values in a dataframe
df=pd.DataFrame({'Actual':y_test,'Predicted':predicted_values})
df
```

```python
# Evaluate Model's Accuracy
accuracy_score=model.score(X_test,y_test)
print("Model's Accuracy:",accuracy_score)
```
